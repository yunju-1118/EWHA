EDA Í≥ºÏ†ïÏóêÏÑú ÌååÏÉùÎ≥ÄÏàòÎ•º ÎßåÎì§Í≥† grouping Ìï¥Î≥¥Í∏∞!:

üß™ Ïó∞Ïäµ Îç∞Ïù¥ÌÑ∞: sales_df

```python
import pandas as pd

sales_df = pd.DataFrame({
    'customer_id': [101, 102, 103, 104, 105, 106],
    'region': ['North', 'South', 'North', 'East', 'East', 'South'],
    'product': ['A', 'B', 'A', 'B', 'A', 'B'],
    'price': [12000, 8000, 15000, 7000, 11000, 9500],
    'quantity': [2, 3, 1, 4, 2, 1]
})
```

‚ùì Î¨∏Ï†ú 1: ÌååÏÉùÎ≥ÄÏàò ÏÉùÏÑ±
priceÏôÄ quantityÎ•º Ïù¥Ïö©ÌïòÏó¨ total_salesÎùºÎäî ÌååÏÉùÎ≥ÄÏàòÎ•º ÎßåÎì§Í≥†, sales_dfÏóê Ï∂îÍ∞ÄÌïòÏãúÏò§.

üîπ ÌûåÌä∏: total_sales = price * quantity

```python
sales_df['total_sales'] = sales_df['price'] * sales_df['quantity']
sales_df
```

‚ùì Î¨∏Ï†ú 2: Ï°∞Í±¥ Í∏∞Î∞ò ÌååÏÉùÎ≥ÄÏàò ÏÉùÏÑ±
Í≥†Í∞ùÏùò total_salesÍ∞Ä 25,000Ïõê Ïù¥ÏÉÅÏù¥Î©¥ 'VIP',
15,000Ïõê Ïù¥ÏÉÅÏù¥Î©¥ 'Regular',
Í∑∏ Ïù¥ÌïòÎäî 'New'Î°ú Íµ¨Î∂ÑÌïòÎäî customer_grade Ïó¥ÏùÑ Ï∂îÍ∞ÄÌïòÏãúÏò§.



```python
def grade(x):
  if x >= 25000:
    return ' VIP'
  elif x>=15000:
    return 'Regular'
  else:
    return 'New'

sales_df['customer_grade'] = sales_df['total_sales'].apply(grade)
```

```python
sales_df
```

‚ùì Î¨∏Ï†ú 3: Í∑∏Î£πÌôî(GroupBy)

Í∞Å regionÎ≥ÑÎ°ú total_salesÏùò Ï¥ùÌï©ÏùÑ Íµ¨ÌïòÏãúÏò§.

Í∞Å productÎ≥ÑÎ°ú quantityÏùò ÌèâÍ∑†ÏùÑ Íµ¨ÌïòÏãúÏò§.

Í∞Å customer_gradeÎ≥ÑÎ°ú Í≥†Í∞ù Ïàò(count)ÏôÄ total_sales ÌèâÍ∑†ÏùÑ Íµ¨ÌïòÏãúÏò§.

region

```python
sales_df.groupby('region')['total_sales'].sum()
```

product

```python
sales_df.groupby('product')['quantity'].mean()
```

customer_grade

```python
sales_df.groupby('customer_grade').count()['customer_id']
```

```python
sales_df.groupby('customer_grade')['total_sales'].mean()
```

üìä Îç∞Ïù¥ÌÑ∞: orders_df (Í∞ÄÏÉÅÏùò Ïò®ÎùºÏù∏ ÏáºÌïëÎ™∞ Ï£ºÎ¨∏ Îç∞Ïù¥ÌÑ∞)

```python
import pandas as pd

orders_df = pd.DataFrame({
    'order_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'customer_id': [101, 102, 101, 103, 104, 105, 104, 106],
    'product': ['A', 'B', 'A', 'C', 'A', 'B', 'C', 'B'],
    'order_date': ['2023-01-01', '2023-01-03', '2023-01-10', '2023-01-12',
                   '2023-01-20', '2023-01-25', '2023-01-28', '2023-01-30'],
    'price': [10000, 8000, 10000, 12000, 10000, 8000, 12000, 8000],
    'quantity': [1, 2, 3, 1, 2, 1, 2, 1]
})
orders_df['order_date'] = pd.to_datetime(orders_df['order_date'])  # ÎÇ†ÏßúÌòïÏúºÎ°ú Î≥ÄÌôò
```

‚ùì Î¨∏Ï†ú 1: ÌååÏÉùÎ≥ÄÏàò ÏÉùÏÑ±
total_price = price * quantityÎùºÎäî ÌååÏÉùÎ≥ÄÏàòÎ•º ÏÉùÏÑ±ÌïòÍ≥† orders_dfÏóê Ï∂îÍ∞ÄÌïòÏãúÏò§.

```python
orders_df['total_price'] = orders_df['price']*orders_df['quantity']
orders_df
```

‚ùì Î¨∏Ï†ú 2: Í≥†Í∞ùÎ≥Ñ ÎàÑÏ†Å Íµ¨Îß§ Í∏àÏï°
customer_idÎ≥Ñ total_priceÏùò Ìï©Í≥ÑÎ•º Íµ¨ÌïòÏãúÏò§.

ÎàÑÏ†Å Íµ¨Îß§Í∏àÏï°Ïù¥ 30000 Ïù¥ÏÉÅÏù∏ Í≥†Í∞ùÎßå Ï∂îÏ∂úÌïòÏãúÏò§.

```python
orders_df.groupby('customer_id')[['total_price']].sum().query('total_price>=30000')
```

‚ùì Î¨∏Ï†ú 3: ÏÉÅÌíàÎ≥Ñ ÌåêÎß§ ÏàòÎüâ Ï†ïÎ†¨
productÎ≥Ñ Ï¥ù ÌåêÎß§ ÏàòÎüâ(quantity)ÏùÑ Íµ¨ÌïòÍ≥†, ÌåêÎß§ ÏàòÎüâ Í∏∞Ï§ÄÏúºÎ°ú ÎÇ¥Î¶ºÏ∞®Ïàú Ï†ïÎ†¨ÌïòÏãúÏò§.

```python
orders_df.groupby('product')['quantity'].sum().sort_values(ascending=True)
```
