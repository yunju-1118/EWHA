##### b-2) Slicing

image.png

slicing -> a[start : stop : step]

````python
# list slicing example
print('sample_list:', sample_list, '\n')

# 1:2는 1부터 (2-1)까지 인덱싱을 하므로 sample_list[1]만을 출력
a = sample_list[1:2]
print(f'sample_list[1:2]: {a}')

# 모든 item을 출력
a = sample_list[:]
print(f'sample_list[:]: {a}')

# sample_list[-4]부터 끝까지 출력
a = sample_list[-4:]
print(f'sample_list[-4:]: {a}')

# 처음부터 sample_list[-4]까지 출력
a = sample_list[:-3]
print(f'sample_list[:-3]: {a}')

# 0,2,4번째 인덱스에 해당하는 값을 출력
a = sample_list[0:5:2]
print(f'sample_list[0:5:2]: {a}')

# sample_list[2]를 기준으로 오른쪽의 모든 값들을 출력
a = sample_list[2::]
print(f'sample_list[2::]: {a}')

# sample_list[2]를 기준으로 왼쪽의 모든 값들의 순서를 뒤집어서 출력
a = sample_list[2::-1]
print(f'sample_list[2::-1]: {a}')

# sample_list[-4]를 기준으로 왼쪽의 모든 값들을 출력
a = sample_list[:-3:]
print(f'sample_list[:-3:]: {a}')

# sample_list[-2]를 기준으로 오른쪽의 모든 값들의 순서를 뒤집어서 출력
a = sample_list[:-3:-1]
print(f'sample_list[:-3:-1]: {a}')
````

## **5. Function**

syntax

```
def your_function_name(parameters):
    statements
    return expression
```





````python
# Ex 1

def mysum1(a,b):
  res = a+b
  return res

mysum1(1,2)
````

````python
# Ex 2

def mysum(n):
  res = 0
  for i in range(0,n+1):
    res = res+i
  return res

mysum(10)
````

여러 개의 입력값을 받는 함수를 만드는 경우 매개변수 이름 앞에 *을 붙이면 입력값을 전부 모아 튜플로 만들어준다.

````python
# Ex 3

def mysum3(*args):
  res = 0
  for i in args:
    res = res+i
  return res

mysum3(1,3,5,7)
````
